```python
import pandas as pd

feature_names = ['party', 'handicapped-infants', 'water-project-cost-sharing',
                 'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                 'el-salvador-aid', 'religious-groups-in-schools',
                 'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                 'mx-missile','immigration', 'synfuels-corporation-cutback',
                 'education-spending', 'superfund-right-to-sue', 'crime',
                 'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv(r'C:\Users\vish8\OneDrive\Desktop\Cursos\RecommendationSystem\DeepLearningIntro\DeepLearningIntro\house-votes-84.data.txt', na_values=['?'], names = feature_names)
voting_data.head()
```

```python
voting_data.dropna(inplace = True)
voting_data.describe()
```

```python
#Normalização dos numeros
voting_data.replace(('y', 'n'), (1, 0), inplace=True)
voting_data.replace(('democrat', 'republican'), (1, 0), inplace=True)
```

```python
all_features = voting_data[feature_names].drop('party', axis=1).values
all_classes = voting_data['party'].values
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

# Criando o modelo
def create_model():
    model = Sequential()
    model.add(Dense(32, input_dim=16, kernel_initializer='he_normal', activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(16, kernel_initializer='he_normal', activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, kernel_initializer='glorot_normal', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Criando estimador compatível com Scikit-Learn
estimator = KerasClassifier(model=create_model, epochs=100, batch_size=32, verbose=0)

# Simulando dados fictícios para testes (remova se já tiver os dados)
np.random.seed(42)
all_features = np.random.rand(1000, 16)  # 1000 amostras, 16 features
all_classes = np.random.randint(0, 2, 1000)  # Classes binárias (0 ou 1)

# Executando validação cruzada
cv_scores = cross_val_score(estimator, all_features, all_classes, cv=10, scoring='accuracy')

# Exibindo média dos scores
print(f"Média da Acurácia: {cv_scores.mean():.4f}")
```
